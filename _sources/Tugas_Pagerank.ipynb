Pertama, lakukan impor pustaka NetworkX dengan menggunakan alias 'nx'.

Selanjutnya, dibuat objek graf 'G' dengan tipe Directed Graph (DiGraph) menggunakan fungsi 'nx.DiGraph()'.

```python
import networkx as nx
G = nx.DiGraph()
```

Kemudian, beberapa edge atau sisi ditambahkan ke graf 'G' menggunakan fungsi 'add_edges_from()'. Edge-edge tersebut merepresentasikan hubungan atau keterkaitan antar node atau simpul pada graf.

Setelah itu, graf 'G' ditampilkan secara visual menggunakan fungsi 'nx.draw()' dengan parameter 'with_labels=True' untuk menampilkan label pada simpul-simpul graf.


```python
G.add_edges_from([(1,2),(2,3),(1,3),(1,4),(2,4),(3,4),(3,5),(4,5),(5,6),(6,7),(6,8),(7,8),(7,9),(8,9),(9,6)])
nx.draw(G, with_labels=True)
```

Pada baris selanjutnya, objek graf 'G' dikonversi menjadi matriks adjacency menggunakan fungsi 'nx.adjacency_matrix()'. Kemudian, hasilnya ditampilkan dalam bentuk matriks dense (array dua dimensi) dengan menggunakan fungsi 'todense()'.

```python
A = nx.adjacency_matrix(G)
A.todense()
```

Melakukan perhitungan nilai PageRank pada graf 'G' menggunakan fungsi 'nx.pagerank()'. PageRank adalah algoritma ranking simpul pada graf yang dikembangkan oleh Larry Page dan Sergey Brin dari Google. Hasil perhitungan PageRank pada graf 'G' disimpan dalam variabel 'pagerank'.

```python
pagerank = nx.pagerank(G)
pagerank
```
