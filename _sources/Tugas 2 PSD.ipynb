## Get Data Sepal-Length from Posgresql Local



*   koneksikan database postgresql menggunakan library psycopg2
*   Setelah itu, sebuah cursor (penunjuk posisi pada database) dibuat untuk menjalankan query SQL.
*   Hasil query disimpan pada variabel hasil_query, kemudian dimasukkan ke dalam DataFrame dengan nama kolom yang diinginkan, yaitu "id" dan "sepallength"
*   Setelah selesai, cursor dan koneksi ditutup menggunakan fungsi close().





```python
import psycopg2
import pandas as pd

# Buat koneksi ke PostgreSQL.
conn = psycopg2.connect(database="prosain200411100159", user="postgres", password="yayanmnh8", host="localhost", port="5432")


# Buat cursor untuk menjalankan query SQL.
cur = conn.cursor()

# Jalankan query SQL.
cur.execute("SELECT ROW_NUMBER() OVER () as id, sepallength FROM iris;")

# Ambil hasil query.
hasil_query = cur.fetchall()

df_sepallength = pd.DataFrame(hasil_query, columns=["id", "sepallength"])

print(df_sepallength)

# Tutup cursor dan koneksi.
cur.close()
conn.close()
```

## Get Data Petal-Length from Mysql Local

*   koneksikan database mysql local menggunakan library mysql.connector
*   Setelah itu, sebuah cursor (penunjuk posisi pada database) dibuat untuk menjalankan query SQL.
*   Hasil query disimpan pada variabel hasil_query, kemudian dimasukkan ke dalam DataFrame dengan nama kolom yang diinginkan, yaitu "id" dan "petallength"
*   Setelah selesai, cursor dan koneksi ditutup menggunakan fungsi close().

```python
import mysql.connector as mc
import pandas as pd

conn = mc.connect(
  host="localhost",
  user="root",
  password="",
  database="prosain200411100159"
)
# Buat cursor untuk menjalankan query SQL.
cur = conn.cursor()

# Jalankan query SQL.
cur.execute("SELECT * FROM iris_arff;")

# Ambil hasil query.
hasil_query = cur.fetchall()

df_petallength = pd.DataFrame(hasil_query, columns=["id", "petallength"])

print(df_petallength)

# Tutup cursor dan koneksi.
cur.close()
conn.close()
```

melakukan penggabungan empat kolom data yang ada dalam beberapa kerangka data (dataframe) menjadi satu dataframe tunggal.

```python
result = pd.concat([df_petallength, df_sepallength, df_sepalwidth, df_petalwidth], axis=1)
result
```



*   menghapus kolom id
*   menyisakan kolom yang di perlukan



```python
result.drop(result.columns[[0,2,4,6]],axis=1,inplace=True)
result
```

```python
X = result.values[:, :4]
Y = result.values[:, 4]
print(X)
```

untuk membagi dataset menjadi data training dan data testing



```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( 
    X, Y, test_size = 0.2, random_state = 100)
```

 implementasi dari algoritma Decision Tree untuk klasifikasi menggunakan Python dan library Scikit-learn (sklearn).

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

Menampilkan akurasi dari metode decision tree

```python
from sklearn.metrics import accuracy_score
print (" DecisionTree Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```

implementasi dari KNN classifier untuk klasifikasi menggunakan Python dan library Scikit-learn (sklearn). 

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

Menampilkan akurasi dari metode KNN Classifier

```python
print (" KNN Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
```
